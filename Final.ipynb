```python
import pandas as pd
import numpy as np

from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv('dataset.csv')
df.head()
```

```python
mean1=df['Latitude'].mean()
mean2=df['Longitude'].mean()
```

```python
df['Latitude'].fillna(value=mean1, inplace=True)
df['Longitude'].fillna(value=mean2, inplace=True)
```

```python
dummies = pd.get_dummies(df['Primary Type'])
df = pd.concat([df,dummies],axis='columns')
df.head()
```

```python
le = LabelEncoder()
```

```python
dfle = df
dfle['Location Description'] = le.fit_transform(dfle['Location Description'])
dfle.head()
```

```python
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
df['Date'] = pd.to_datetime(df['Date'], format = '%m/%d/%Y %H:%M:%S %p')
```

```python
column_1 = df.iloc[:,0]

db=pd.DataFrame({
              "day": column_1.dt.day,
              "hour": column_1.dt.hour,
              "dayofweek": column_1.dt.dayofweek,
             })
```

```python
df=pd.concat([db,df],axis=1)
df.drop(['Date','Block','Primary Type','Block'],axis=1,inplace = True)
df.dropna(inplace=True)
```
